```python
df=pd.read_csv('../Data/cleanTweets.csv',nrows=500000).fillna('')
```

```python
from gensim.models.phrases import Phrases, Phraser
phrases = Phrases(sentences=df['tokens'],min_count=10,threshold=5)
bigram = Phraser(phrases)
```

```python
s='sachin pilot is a leader'
print(bigram[s.split()])
```

```python
#below we are modifying tokens in order to replace popular bigrams with single token eg ['rahul','gandhi'] becomes
#['rahul_gandhi']
def f(arg):
    return bigram[arg['tokens']]


df['tokens']=df.apply(f,axis=1)
df=shuffle(df)
df=df.reset_index(drop=True)
```

```python
print(df['tokens'])
```

```python
from gensim.models import Word2Vec
#import multiprocessing
#cores = multiprocessing.cpu_count()
tokens=df['tokens']
#print(cores)

t=time()
obj_w2v= Word2Vec(sentences=tokens, min_count=20,window=2,size=300,sample=6e-5, alpha=0.03, min_alpha=0.0007, negative=20,workers=3,iter=30)

print(time()-t)
```

```python
print (list(obj_w2v.wv.vocab) )
```
